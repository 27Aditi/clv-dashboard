```python
customer_df = df.groupby('Customer ID').agg(
    FirstPurchaseDate = ('InvoiceDate', 'min'),
    LastPurchaseDate = ('InvoiceDate', 'max'),
    NumofTransactions = ('Invoice', 'nunique'),
    Monetary = ('TotalAmount', 'sum')
)
```

```python
from datetime import datetime
def calculateRecency(customer_df):
    analysis_date = datetime(2011, 12, 10, 0, 0, 0)
    return analysis_date - customer_df['LastPurchaseDate']
```

```python
customer_df['Recency'] = calculateRecency(customer_df)
```

```python
customer_df.head()
```

```python
customer_df['Recency'] = customer_df['Recency'].dt.days
```

```python
customer_df.head()
```

```python
customer_df.info()
```

```python
customer_df = customer_df.rename(columns = {'NumofTransactions' : 'Frequency'})
```

```python
customer_df.head()
```

```python
def getaverage(customer_df):
    return customer_df['Monetary'] / customer_df['Frequency']
```

```python
customer_df['AvgOrderValue'] = getaverage(customer_df)
```

```python
customer_df.head()
```

```python
customer_df['HistoricalCLV'] = customer_df['Monetary']
```

```python
customer_df['M_Score'] = pd.qcut(customer_df['Monetary'], q=5, labels=[1,2,3,4,5])
```

```python
customer_df.head()
```

```python
customer_df['R_Score'] = pd.qcut(customer_df['Recency'], q=5, labels=[5,4,3,2,1])
```

```python
customer_df.head()
```

```python
customer_df['F_Score'] = pd.qcut(customer_df['Frequency'].rank(method='first'), q=5, labels=[1,2,3,4,5])
```

```python
# Qcut does not work on duplicate values
# Uses rank columns to number rows even when values are same
```

```python
customer_df.head()
```

```python
customer_df['CustomerLifetime'] = customer_df['LastPurchaseDate'] - customer_df['FirstPurchaseDate']
```

```python
customer_df.head()
```

```python
customer_df['CustomerLifetime'] = customer_df['CustomerLifetime'].dt.days
```

```python
customer_df.head()
```

```python
customer_df['CustomerLifetime'] = customer_df['CustomerLifetime'].replace(0, 1)
```

```python
def calculatePurchaseRate(row):
    if(row['CustomerLifetime'] == 0):
        row['CustomerLifetime'] = 1
        return 0
    return row['Frequency'] / row['CustomerLifetime']
```

```python
customer_df['PurchaseRate'] = customer_df.apply(calculatePurchaseRate, axis=1)
```

```python
customer_df.head()
```

```python
def calculatePredictedPurchases(row):
    return row['PurchaseRate'] * 365
```

```python
customer_df['PredictedPurchases'] = customer_df.apply(calculatePredictedPurchases, axis=1)
```

```python
customer_df.head()
```

```python
def calculatePredictedCLV(row):
    return row['PredictedPurchases'] * row['AvgOrderValue']
```

```python
customer_df['PredictedCLV'] = customer_df.apply(calculatePredictedCLV, axis=1)
```

```python
customer_df.head()
```

```python
customer_df.info()
```

```python
customer_df['TotalCLV'] = customer_df['HistoricalCLV'] + customer_df['PredictedCLV']
```

```python
customer_df.head()
```

```python
customer_df.describe()
```

```python
def calculateSegment(row):
    if (row['R_Score'] >= 4 and row['F_Score'] >= 4 and row['M_Score'] >= 4):
        return 'Premium'
    elif (row['F_Score'] >= 4 and row['M_Score'] >= 4):
        return 'Core'
    elif (row['R_Score'] >= 4 and (row['F_Score'] >= 2 or row['M_Score'] >= 2)):
        return 'Promising'
    elif (row['R_Score'] >= 4 and row['F_Score'] <= 2):
        return 'Emerging'
    elif (row['R_Score'] <= 2 and row['F_Score'] >= 3):
        return 'Declining'
    elif (row['R_Score'] <= 2 and row['M_Score'] >= 4):
        return 'Critical'
    elif (row['R_Score'] <= 2 and row['M_Score'] <= 2 and row['F_Score'] <= 2):
        return 'Inactive'
    else:
        return 'Watchlist'
```

```python
customer_df['Segment'] = customer_df.apply(calculateSegment, axis=1)
```

```python
customer_df.head()
```

```python
customer_df.info()
```

```python
customer_df.to_parquet('customer_wise')
```

```python
Segment_df = customer_df.groupby('Segment').agg(
    Count = ('Customer ID', 'count'),
    TotalCLV = ('TotalCLV', 'sum'),
    AverageCLV = ('TotalCLV', 'mean'),
    AvgRecency = ('Recency', 'mean'),
    AvgFrequency = ('Frequency', 'mean'),
    AvgMonetary = ('Monetary', 'mean')
)
```
